```python
### Data collection and analysis

import pandas as pd
import numpy as np

customer_data=pd.read_csv('Mall_Customers.csv')
customer_data

```

```python
### x values 
x=customer_data.iloc[:,3:5].values
x
```

```python
### preprocessing

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
### find the kmeans

from sklearn.cluster import KMeans
k=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,init='k-means++')
    km.fit(x)
    k.append(km.inertia_)
k
```

```python
### plot the ELBOW
import matplotlib.pyplot as plt
plt.plot(k)
plt.title('The ELBOW point Graph')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## no of clusters predict
km=KMeans(n_clusters=5,init='k-means++')
a=km.fit_predict(x)
a
```

```python
customer_data['result']=pd.DataFrame(a)
customer_data
```

```python
### plot the clusters 
plt.scatter(x[a==0,0],x[a==0,1],s=75,c='red')
plt.scatter(x[a==1,0],x[a==1,1],s=75,c='pink')
plt.scatter(x[a==2,0],x[a==2,1],s=75,c='green')
plt.scatter(x[a==3,0],x[a==3,1],s=75,c='yellow')
plt.scatter(x[a==4,0],x[a==4,1],s=75,c='violet')

###plot the centroid 

plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],s=100,c='black',label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()

```
